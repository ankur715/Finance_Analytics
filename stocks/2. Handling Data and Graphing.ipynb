```python
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas as pd
import pandas_datareader.data as web
```

```python
style.use('ggplot')

start = dt.datetime(2015, 1, 1)
end = dt.datetime.now()
df = web.DataReader("TSLA", 'yahoo', start, end)
df.reset_index(inplace=True)
df.set_index("Date", inplace=True)
```

```python
df.to_csv('TSLA.csv')
```

Rather than reading data from Yahoo's finance API to a DataFrame, we can also read data from a CSV file into a DataFrame:



```python
df = pd.read_csv('tsla.csv', parse_dates=True, index_col=0)
```

```python
df.plot()
plt.show()
```

Cool, except that the only thing we can really see here is the volume, since it's on a scale MUCH larger than stock price.

```python
df['Adj Close'].plot()
plt.show()
```

As you can see, you can reference specific columns in the DataFrame like: df['Adj Close'], but you can also reference multiple at a time, like so:

```python
df[['High','Low']]
```
